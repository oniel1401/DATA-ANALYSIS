```python
!pip install feedparser
import os
import time
import feedparser
import pandas as pd

# URLs des flux RSS
urls = {
    "societe": "https://www.francetvinfo.fr/societe.rss",
    "science": "https://www.francetvinfo.fr/sciences.rss",
    "economie": "https://www.francetvinfo.fr/economie.rss"
}

#Stockage
csv_filename = "news_data.csv"

#collectede donnéess
def collect(url):
    feed = feedparser.parse(url)
    articles = []
    for entry in feed.entries:
        articles.append({"title": entry.title, "summary": entry.summary, "published": entry.published, "categorie": categorie })
    return articles

# save dans le csv
def save(data):
    # créé le ficher
    if not os.path.isfile(csv_filename):
        df = pd.DataFrame(data)
        df.to_csv(csv_filename, index=False)
        print("Nouveau fichier CSV créé et données enregistrées.")
    else:
        # load data
        existing_df = pd.read_csv(csv_filename)
        existing_titles = existing_df['title'].tolist()

        # titre exist
        new_data = [article for article in data if article['title'] not in existing_titles]

        # add
        if new_data:
            new_df = pd.DataFrame(new_data)
            new_df.to_csv(csv_filename, mode='a', header=False, index=False)
            print(f"{len(new_data)} nouvelles données ajoutées ")
        else:
            print("Aucune donnée à ajouter.")

while True:
    for categorie, url in urls.items():
        articles = collect(url)
        save(articles)

    time.sleep(300)
```
